```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_sample = pd.read_excel('/home/centos/leon/machine_learning_jupyter/dataset/collection_speech_sample.xlsx')
print('样本总数：{}, 正样本数：{}，负样本数：{}'.format(df_sample.shape[0], df_sample[df_sample['y'] == 1].shape[0], df_sample[df_sample['y'] == 0].shape[0]))
df_sample.head()
```

#### 正样本的正面、中性、负面的情况

```python
df_sample_positive = df_sample[df_sample['y'] == 1]
df_sample_positive_optimistic = df_sample_positive.iloc[:, 0:11]  # 正面
df_sample_positive_neutral = df_sample_positive.iloc[:, 11:22]  # 中性
df_sample_positive_pessimistic = df_sample_positive.iloc[:, 22:33]  # 负面
```

```python
pd.DataFrame({'正面': df_sample_positive_optimistic.sum().values[1:], 
              '中性': df_sample_positive_neutral.sum().values[1:], 
              '负面': df_sample_positive_pessimistic.sum().values[1:]}, 
             columns=['正面', '中性', '负面'], 
             index=['{} 分布值'.format(v/10.0) for v in range(1, 11)])
```

#### 负样本的正面、中性、负面的情况

```python
df_sample_negative = df_sample[df_sample['y'] == 0]
df_sample_negative_optimistic = df_sample_negative.iloc[:, 0:11]  # 正面
df_sample_negative_neutral = df_sample_negative.iloc[:, 11:22]  # 中性
df_sample_negative_pessimistic = df_sample_negative.iloc[:, 22:33]  # 负面
```

```python
pd.DataFrame({'正面': df_sample_negative_optimistic.sum().values[1:], 
              '中性': df_sample_negative_neutral.sum().values[1:], 
              '负面': df_sample_negative_pessimistic.sum().values[1:]}, 
             columns=['正面', '中性', '负面'], 
             index=['{} 分布值'.format(v/10.0) for v in range(1, 11)])
```
